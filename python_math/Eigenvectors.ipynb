```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Let's start with a simple linear transformation
A = np.array([[2, 1], [1, 2]])

# Let's show how this transforms the points (0, 1) and (1, 0)
plt.figure(figsize=(10, 8))
plt.scatter(0, 1, color = 'blue')
plt.scatter(1, 0, color = 'red')
plt.scatter(-1, 0, color = 'green')
plt.scatter(0, -1, color = 'black')
axes = plt.gca()
axes.set_xlim([-2.2, 2.2])
axes.set_ylim([-2.2, 2.2])
plt.show()

B = np.matmul([[0, 1], [1, 0], [-1, 0], [0, -1]], A)
plt.figure(figsize=(10, 8))
plt.scatter(B[0, 0], B[0, 1], color = 'blue', alpha = 0.8)
plt.scatter(B[1, 0], B[1, 1], color = 'red', alpha = 0.8)
plt.scatter(B[2, 0], B[2, 1], color = 'green', alpha = 0.8)
plt.scatter(B[3, 0], B[3, 1], color = 'black', alpha = 0.8)
axes = plt.gca()
axes.set_xlim([-2.2, 2.2])
axes.set_ylim([-2.2, 2.2])
plt.show()
```

## Finding Eigenvalues with the Characteristic Polynomial
Here is the characteristic polynomial
det(A - lambda*I) = 0 
lambda ** 2 - 4 * lambda + 3 = 0
lambda1 = 1
lambda2 = 3

## Now for the Eigenvectors
(A - lambda1 * I) * [v1, v2] = 0
his implies
v1 + v2 = 0
so eigenvector1 = [1, -1]

Now calculate the second eigenvector using the second eigenvalue
(A - lambda2 * I) * [v1, v2] = 0
this implies
- v1 + v2 = 0
so eigenvector2 = [1, 1]

```python
# The eigenvectors are those vectors whose direction is preserved 
# through the transformation. I've plotted the eigenvectors in 
# yellow below. The eigenvalues corresponding to the eigenvectors
# are the scalar multiples applied to the vectors during the 
# transformation.

plt.figure(figsize=(10, 8))
plt.scatter(0, 1, color = 'blue')
plt.scatter(1, 0, color = 'red')
plt.scatter(-1, 0, color = 'green')
plt.scatter(0, -1, color = 'black')
plt.scatter(1, 1, color = 'yellow')
plt.scatter(1, -1, color = 'yellow')
axes = plt.gca()
axes.set_xlim([-3.2, 3.2])
axes.set_ylim([-3.2, 3.2])
plt.show()

B = np.matmul([[0, 1], [1, 0], [-1, 0], [0, -1], [1, 1], [1, -1]], A)
plt.figure(figsize=(10, 8))
plt.scatter(B[0, 0], B[0, 1], color = 'blue', alpha = 0.8)
plt.scatter(B[1, 0], B[1, 1], color = 'red', alpha = 0.8)
plt.scatter(B[2, 0], B[2, 1], color = 'green', alpha = 0.8)
plt.scatter(B[3, 0], B[3, 1], color = 'black', alpha = 0.8)
plt.scatter(B[4, 0], B[4, 1], color = 'yellow', alpha = 0.8)
plt.scatter(B[5, 0], B[5, 1], color = 'yellow', alpha = 0.8)
axes = plt.gca()
axes.set_xlim([-3.2, 3.2])
axes.set_ylim([-3.2, 3.2])
plt.show()
```

```python
# Sometimes eigenvalues and eigenvectors can be complex. Consider the following
# case, where there is a rotation by 90 degrees. All vectors change direction
# with this transformation.

A = np.array([[0, -1], [1,0]])

# Let's show how this transforms the points (0, 1) and (1, 0)
plt.figure(figsize=(10, 8))
plt.scatter(0, 1, color = 'blue')
plt.scatter(1, 0, color = 'red')
plt.scatter(-1, 0, color = 'green')
plt.scatter(0, -1, color = 'black')
axes = plt.gca()
axes.set_xlim([-2.2, 2.2])
axes.set_ylim([-2.2, 2.2])
plt.show()

B = np.matmul([[0, 1], [1, 0], [-1, 0], [0, -1]], A)
plt.figure(figsize=(10, 8))
plt.scatter(B[0, 0], B[0, 1], color = 'blue', alpha = 0.5)
plt.scatter(B[1, 0], B[1, 1], color = 'red', alpha = 0.5)
plt.scatter(B[2, 0], B[2, 1], color = 'green', alpha = 0.5)
plt.scatter(B[3, 0], B[3, 1], color = 'black', alpha = 0.5)
axes = plt.gca()
axes.set_xlim([-2.2, 2.2])
axes.set_ylim([-2.2, 2.2])
plt.show()
```
