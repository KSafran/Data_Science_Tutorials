```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# First, let's define some linear transformation from R2 to R2
# AX = Y

A = np.array([[1, 2], [-1, 0.5]])

# Let's show how this transforms the points (0, 1) and (1, 0)
plt.figure(figsize=(10, 8))
plt.scatter(0, 1, color = 'blue')
plt.scatter(1, 0, color = 'red')
axes = plt.gca()
axes.set_xlim([-1.2, 1.2])
axes.set_ylim([-1.2, 1.2])
plt.show()

B = np.matmul(A, np.transpose([[0, 1], [1, 0]]))
print(B)
plt.figure(figsize=(10, 8))
plt.scatter(0, 1, color = 'blue')
plt.scatter(1, 0, color = 'red')
axes = plt.gca()
axes.set_xlim([-1.2, 1.2])
axes.set_ylim([-1.2, 1.2])
plt.show()
```
