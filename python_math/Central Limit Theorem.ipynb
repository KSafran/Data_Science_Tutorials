```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
%matplotlib inline
```

# Central Limit Theorem
The central limit theorem (CLT) states that the mean of a big enough sample from a reasonably 
well-behaved distribution follows a normal distribution with mean $\mu$ and variance  $\sigma^2/\sqrt(n)$.
Here $\mu$ is the population mean, $\sigma^2$ is the population variance, and n is the size of each sample.

(i.e. sample means from (almost) any distribution are normally distributed. Examples below...)

## Normal Distribution
Let's try this out with a normal distriubtion. We will observe the mean of samples of 10 from a normal distribution 
with $\mu$ = 0 and $\sigma^2$ = 1.

```python
normal_means = []
for i in np.arange(1000):
   normal_means.append(np.mean(np.random.normal(size = 10)))
```

```python
# These values are normally distributed
plt.hist(normal_means)
```

```python
stats.probplot(normal_means, dist = "norm", plot = pylab) 
pylab.show()
```

```python
# and the mean and variance agree with the CLT 
print(format('expected mean = 0, actual mean = %.2f' % np.mean(normal_means))) 
print(format('expected variance = 1/10, actual variance = %.2f' % np.var(normal_means)))
```

## Uniform Distribution 
Maybe less intuitively, this also works with a uniform distribution.

```python
uniform_means = [] 
for i in np.arange(1000): 
    uniform_means.append(np.mean(np.random.uniform(size = 10))) 
```

```python
# These values are also normally distributed 
plt.hist(uniform_means)
```

```python
stats.probplot(uniform_means, dist = "norm", plot = pylab)
pylab.show()
```

```python
print(format('expected mean = 0.5, actual mean = %.2f' % np.mean(uniform_means))) 
print(format('expected variance = %.4f, actual variance = %.4f' % (1/120, np.var(uniform_means))))
```

## Poisson Distribution

```python
poisson_means = []
for i in np.arange(1000):
    poisson_means.append(np.mean(np.random.poisson(size = 10)))
```

```python
# Again, these are normal
plt.hist(poisson_means)
```

```python
stats.probplot(poisson_means, dist = "norm", plot = pylab)
pylab.show()
```

```python
print(format('expected mean = 1, actual mean = %.2f' % np.mean(poisson_means)))
print(format('expected variance = 1/10, actual variance = %.4f' % np.var(poisson_means)))
```
