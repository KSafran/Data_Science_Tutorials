```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Take a look at the graph of ln(x)

x = np.arange(0.1, 100, 0.1)
y = np.log(x)

plt.figure(figsize=(10, 8)) 
plt.plot(x, y, 'r-', color = 'blue')
plt.show()
```

```python
# Now let's approximate ln(x) with a taylor series

# Taylor series follows the following rule
# f(a) + f'(a) * (x - a)/1! + f"(a) * f(a) * (x - a)/2!  + ... + 

# We will evaluate this for ln(x) at a = 1, starting with the first term
# t1 = f(a) = ln(1) = 0
t1 = np.repeat(0, np.size(x))

plt.figure(figsize=(10, 8)) 
plt.plot(x, y, 'r-', color = 'blue')
plt.plot(x, t1, 'r--', color = 'red')

plt.show()
```

```python
# Now the second term
# t2 = f(a) + f'(a) * (x - a) = ln(1) + 1/1 * (x - 1) = x - 1
t2 = x - 1

plt.figure(figsize=(10, 8)) 
plt.plot(x, y, 'r-', color = 'blue')
plt.plot(x, t2, 'r--', color = 'red')
axes = plt.gca()
axes.set_ylim([-2.5, 5])
plt.show()
```

```python
# Now the third
# t2 = f(a) + f'(a) * (x - a) + f"(a) * (x - a)**2/2 = x - 1 - (x - 1)/2
t3 = x - 1 - ((x - 1)**2)/2

plt.figure(figsize=(10, 8)) 
plt.plot(x, y, 'r-', color = 'blue')
plt.plot(x, t3, 'r--', color = 'red')
axes = plt.gca()
axes.set_ylim([-2.5, 5])
plt.show()
```

```python
# Now in general
# tn = sum((-1) ** i * (x - 1)/i!)
taylor = np.repeat(0., np.size(x))
for i in range(1, 10):
    taylor += ((-1) ** (i - 1)) * (x - 1) ** i / (i * np.math.factorial(i))

plt.figure(figsize=(10, 8)) 
plt.plot(x, y, 'r-', color = 'blue')
plt.plot(x, taylor, 'r--', color = 'red')
axes = plt.gca()
axes.set_ylim([-2.5, 5])
plt.show()
```
