```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model, kernel_ridge
%matplotlib inline
```

# Kernel Regression
When there is a clear non-linear relationship between a predictor and some response variable then simple linear regression is inappropriate. You could fit a polynomial to the predictor, or some regularized spline. Another alternative is to use kernel regression. Kernel regression is a non-parametric technique, in which our predictions are a weighted average of the response for each point, where the weight is related to the distance of each point from our predictor.

$$\hat{\theta} = \sum_{i}\frac{  y_i  kernel(x, x_i)}{\sum_{i} kernel(x, x_i)}  $$

```python
np.random.seed(0)
x = np.random.uniform(size = 100, low = 0, high = 1)
x.sort() # so kernel regression line segments will connect appropriately
y = np.sin(3.14 * x) + np.random.normal(size = 100, loc=0, scale = 0.1)
plt.scatter(x, y)
```

## Linear Regression
Below we see the obvious problem with linear regression.

```python
basic_model = linear_model.LinearRegression()
basic_model.fit(X = x.reshape(-1, 1), y = y)
plt.scatter(x, y)
plt.plot(x, basic_model.predict(x.reshape(-1, 1)), color = 'red', linewidth = 2)
plt.show()
```

## Kernel Regression
Kernel regression fixes this by using local points to predict the response. In terms of the bias-variance tradeoff, this model haslower bias, higher variance than the linear regression

```python
kernel_model = kernel_ridge.KernelRidge(alpha = 0.00001, kernel = 'rbf')
kernel_model.fit(X = x.reshape(-1, 1), y = y)
plt.scatter(x, y)
plt.plot(x.reshape(-1, 1), kernel_model.predict(x.reshape(-1, 1)), 
         color = 'red', linewidth = 2)
plt.show()
```
